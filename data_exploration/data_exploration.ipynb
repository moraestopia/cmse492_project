```python
import pandas as pd
import os
```

```python
data_pwd = "../datasets/adidas_tick_options.csv"
```

```python
option_raw = pd.read_csv(data_pwd)
option_raw
```

```python
#######################################
#### Create 2 dfs: Calls and Puts #####
#######################################

# Filter Only EUROPEAN STYLE Options - as we are comparing against Black-Scholes

option_raw_eu = option_raw[option_raw["ExerciseStyle"] == "E"]


# General Calls and Puts
call_df = option_raw[option_raw["CallPut"] == "C"] # Call DataFrame

put_df = option_raw[option_raw["CallPut"] == "P"] # Put DataFrame

# European Calls and Puts
call_eu_df = option_raw_eu[option_raw_eu["CallPut"] == "C"]
put_eu_df = option_raw_eu[option_raw_eu["CallPut"] == "P"]

save_dir = os.path.dirname(data_pwd)


# Paths to save the new DataFrames
call_save_path = os.path.join(save_dir, "call_options.csv")
put_save_path = os.path.join(save_dir, "put_options.csv")

put_eu_save_path = os.path.join(save_dir, "put_options_eu.csv")
call_eu_save_path = os.path.join(save_dir, "call_options_eu.csv")

# Save the DataFrames as CSV
call_df.to_csv(call_save_path, index=False)
put_df.to_csv(put_save_path, index=False)

call_eu_df.to_csv(call_eu_save_path, index=False)
put_eu_df.to_csv(put_eu_save_path, index=False)


print(f"Call DataFrame saved at: {call_save_path}")
print(f"Put DataFrame saved at: {put_save_path}")
print(f"Call DataFrame saved at: {call_eu_save_path}")
print(f"Put DataFrame saved at: {put_eu_save_path}")
```
